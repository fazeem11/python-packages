```python
import pandas as pd
df = pd.read_csv('Wine.csv')
df.head()
```

```python
df.shape
```

```python
wcss=[]
```

```python
df1=df.copy()
df1
```

```python
x=df1[['Alcohol', 'Magnesium', 'Proline']]
x.head()
```

```python
wcss=[]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
import matplotlib.pyplot as plt
plt.plot(wcss,marker='o')
plt.xlabel('x values')
plt.ylabel('wcss')
plt.show
```

```python
wcss
```

```python
km1 = KMeans(n_clusters=3)

km1
```

```python
km1.fit(x)
```

```python
y_pred=km1.predict(x)
y_pred
```

```python
df1['Label']=y_pred
df1.head()
```

```python
import seaborn as sns
sns.scatterplot(x=df1['Alcohol'],y=df1['Color_Intensity'],hue=df1['Label'])
```

```python
from sklearn import metrics
score = metrics.silhouette_score(x,df1["Label"])
score
```
